<h2 style="text-align:center">ScatterPlot</h2>

<h3>A scatter plot is a type of data visualization that displays values for typically two variables for a set of data using Cartesian coordinates. Each point on the scatter plot represents an observation in the dataset, with the position along the x-axis and y-axis representing the values for the two variables.</h3>

```python
price=list(df['price'].values)
age=list(df['age'].values)

# we must pass 2 columns and both columns must be numerical
plt.scatter(age,price,color="red")  ## plotting graph with scatterplot.

plt.xlabel("AGE") ## passing label to x-axis
plt.ylabel("PRICE") ## passing label to y-axis
plt.title("AGE Vs PRICE") ## passing title to graph
plt.show() ## used to show graph.
```

<h2 style="text-align:center">BarPlot</h2>

<h3>A bar plot (or bar chart) is a graphical display of data using rectangular bars to represent different categories of data. The length or height of each bar is proportional to the value it represents.</h3>

```python
brand =list(df['brand'].value_counts().keys())  ## filtering out brand names and storing it into list.
no_of_counts=list(df['brand'].value_counts().values) ## no of counts
plt.figure(figsize=(16,8)) ## passing size of graph
plt.bar=(brand,no_of_counts) ## plotting bar plot with 2 arguments as x and y axis.
plt.xlabel('BRANDS') ## passing label to x-axis
plt.ylabel('no_of_counts') ## passing label to y-axis
plt.title("No. of Bike in Brand") ## passing title to graph
plt.show()
plt.xticks(rotation=60)  
```

<h2 style="text-align:center">Pie Plot</h2>

<h3>A pie plot (or pie chart) is a circular statistical graphic divided into slices to illustrate numerical proportions. Each slice represents a category's contribution to the whole, with the size of the slice proportional to the quantity it represents.</h3>

```python
ownership = list(df['owner'].value_counts().keys()) 
no_of_count = list(df['owner'].value_counts().values)
plt.figure(figsize=(8,8))
plt.pie(no_of_count,labels=ownership,autopct='%1.2f%%') ## plotting pie plot for ownership.
plt.show()
```

```python
country=['India','China','Japan','Pakistan']
population=[900,700,500,350] 
explosion=[0,0.2,0,0.2]
# data
```

<h2 style="text-align:center">Boxplot</h2>

<h3>A boxplot is a graphical representation of a dataset's distribution, showing the minimum, first quartile (Q1), median, third quartile (Q3), and maximum, with outliers often indicated.<h3>

<h4>Components of a Boxplot:-

1) Box: The main part of the plot that shows where the central portion (interquartile range) of the data lies, from Q1 to Q3.

2) Whiskers: Lines extending from the box to the minimum and maximum values within 1.5 * IQR (Interquartile Range, the distance between Q1 and Q3) from Q1 and Q3, respectively.

3) Outliers: Data points outside the range of the whiskers are considered outliers and are often marked with dots or asterisks.</h4>

```python
plt.boxplot(bullet['price']) ## ploting boxplot for price of Royal Enfield
plt.show
```

```python
bullet_price=bullet[(bullet['price']>=5000) & (bullet['price']<=200000)] ## filtering data according to price
```

```python
plt.boxplot(bullet_price['price']) ## plotting box plot for filtered price of Royal Enfield
plt.plot()
```
